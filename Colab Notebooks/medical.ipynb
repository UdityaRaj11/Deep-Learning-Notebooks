### RAG Pipeline: Embedding + Gemini (LLM)

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=GOOGLE_API_KEY)
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever(search_kwargs={"k":5})
```

```python
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say you don't know, don't try to make up an answer. Keep the answers as concise as possible. Answer as a doctor.
{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)
qa_chain = RetrievalQA.from_chain_type(
    model,
    retriever=vector_index,
    return_source_documents=True,
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)
```

```python
question="What are Symptoms of diarrhoea ?"
result=qa_chain({"query": question})
Markdown(result["result"])
```
