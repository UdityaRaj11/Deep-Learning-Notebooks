```python
def build_model(hp):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(units=hp.Int('units', min_value=32, max_value=512, step=32),
                              activation='relu'),
        tf.keras.layers.Dropout(rate=hp.Float('dropout', min_value=0.0, max_value=0.5, step=0.1)),
        tf.keras.layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='mae',
                  metrics=['mse'])
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_mse',
    max_trials=10,
    directory='hyperparameter_tuning03',
    project_name='my_project'
)

tuner.search(X_train, y_train, epochs=10, validation_data=(X_train, y_train))
```

```python
def build_model(hp):
      inputs = layers.Input(shape=(8))
      x = layers.Lambda(lambda x: tf.expand_dims(x, axis=1))(inputs) # expand input dimension to be compatible with LSTM
      # print(x.shape)
      # x = layers.LSTM(128, activation="relu", return_sequences=True)(x) # this layer will error if the inputs are not the right shape
      x = layers.LSTM(units=hp.Int('units', min_value=32, max_value=512, step=32), activation="relu")(x)
      x = layers.Dense(units=hp.Int('units', min_value=32, max_value=512, step=32), activation="relu")(x) # using the tanh loss function results in a massive error
      # print(x.shape)
      # Add another optional dense layer (you could add more of these to see if they improve model performance)
      # x = layers.Dense(32, activation="relu")(x)
      output = layers.Dense(1)(x)
      model_08 = tf.keras.Model(inputs=inputs, outputs=output, name="model_7_lstm")

      # Compile the model
      model_08.compile(loss=tf.keras.losses.mae,
                                optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                                metrics=['mse'])
      return model_08
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_mse',
    max_trials=50,
    directory='hyperparameter_tuning2',
    project_name='my_project'
)

tuner.search(X_train, y_train, epochs=10, validation_data=(X_train, y_train))
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model_07_best = tuner.hypermodel.build(best_hps)
model_07_best.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```
